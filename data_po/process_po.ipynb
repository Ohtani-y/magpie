```python
# add path 
import sys
sys.path.append("../")

from tqdm import tqdm
from exp.utils import *
from datasets import load_dataset, Dataset, DatasetDict

# Please change this file path
all_data = load_dataset_from_file("example_instructions_5res_armorm.json")

for data in tqdm(all_data):
    rewards_armorm = [item['score'] for item in data['rewards_armorm']]
    max_reward_index = rewards_armorm.index(max(rewards_armorm))
    min_reward_index = rewards_armorm.index(min(rewards_armorm))
    max_reward_response = data['responses'][max_reward_index]
    min_reward_response = data['responses'][min_reward_index]

    data['chosen'] = [{"role": "user", "content": data['instruction']}, {"role": "assistant", "content": max_reward_response}]
    data['rejected'] = [{"role": "user", "content": data['instruction']}, {"role": "assistant", "content": min_reward_response}]
```
